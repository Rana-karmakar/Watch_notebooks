```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("W_YFA0002_UID_198D94B1-7CD8-4D3C-BCD3-14C9374A1F3D_TS_12-07-2022-GMT%200530-04-41-14-PM.csv")
df.head()
```

```python
pitch = df['Pitch']
yaw = df['Yaw']
roll = df['Roll']
```

```python
# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(pitch, label='Pitch')
ax.plot(yaw, label='Yaw')
ax.plot(roll, label='Roll')

# Set labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Values')
ax.set_title('Apple Watch Sensor Data')

# Display the legend
ax.legend()

# Show the plot
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
%matplotlib notebook

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Initialize the line object
line, = ax.plot([], [], [])

# Set up the initial frame
def init():
    line.set_data([], [])
    line.set_3d_properties([])
    return line,

# Define the animation update function
def update(frame):
    # Update the line data
    line.set_data(pitch[:frame], yaw[:frame])
    line.set_3d_properties(roll[:frame])
    return line,

# Create the animation
animation = FuncAnimation(fig, update, frames=len(pitch), init_func=init, blit=True)

# Set labels and title
ax.set_xlabel('Pitch')
ax.set_ylabel('Yaw')
ax.set_zlabel('Roll')
ax.set_title('Apple Watch Movement Data Animation')

# Show the animation
plt.show()
```
