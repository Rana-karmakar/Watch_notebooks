```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data/server_data/YFA0013-21-08-2023-GMT 0530-11-06-51-AM.csv")
df.head()
```

```python
pitch = df['Pitch']
yaw = df['Yaw']
roll = df['Roll']
fig, ax = plt.subplots()
ax.plot(pitch, label='Pitch')
ax.plot(yaw, label='Yaw')
ax.plot(roll, label='Roll')
ax.set_xlabel('Time')
ax.set_ylabel('Values')
ax.set_title('Apple Watch Sensor Data')
ax.legend()
plt.show()
```
