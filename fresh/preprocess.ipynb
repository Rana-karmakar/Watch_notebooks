```python
import os
import pandas as pd
```

```python
input_folder = 'C:/Users/ranak/OneDrive/Desktop/my_desk/watch_notebooks/fresh/server_data/currect_data'
output_folder = 'C:/Users/ranak/OneDrive/Desktop/my_desk/watch_notebooks/fresh/server_data/processed_data'
```

```python
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
# Get a list of all CSV files in the input folder
csv_files = [file for file in os.listdir(input_folder) if file.endswith('.csv')]

for csv_file in csv_files:
    df = pd.read_csv(os.path.join(input_folder, csv_file))
    new_column_value = csv_file[:7]

    # Add a new column with the extracted value
    df['asana_id'] = new_column_value

    # Save the modified DataFrame to a new CSV file in the output folder
    new_file_name = f"{new_column_value}.csv"
    output_file = os.path.join(output_folder, new_file_name)
    df.to_csv(output_file, index=False)

    # Rename the output file to the new name
    os.rename(output_file, os.path.join(output_folder, new_file_name))

print("Files processed and saved with new names.")
```

```python
output_files = [file for file in os.listdir(input_folder) if file.endswith('.csv')]
output_files
```

```python
# List to store individual DataFrames
dataframes = []

for output_file in output_files:
    # Read the CSV file
    df = pd.read_csv(os.path.join(input_folder, output_file))

    # Append the DataFrame to the list
    dataframes.append(df)
print(dataframes)
```

```python
# Concatenate all DataFrames into a single DataFrame
concatenated_df = pd.concat(dataframes, ignore_index=True)

# Save the concatenated DataFrame to a single CSV file
concatenated_file_path = os.path.join(output_folder, 'concatenated.csv')
concatenated_df.to_csv(concatenated_file_path, index=False)

print("Files concatenated and saved as concatenated.csv.")
```
